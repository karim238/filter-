```python
from tkinter import *
from tkinter import ttk
from PIL import Image, ImageTk
import cv2
from matplotlib import pyplot as plt
import numpy as np

root=Tk()

root.title("filter")
root.geometry("400x400")




def read_img():
    

    img=cv2.imread(path.get())
    cv2.imshow("original",img)
    cv2.waitKey(0)
    return img



def neg_img():
    img=cv2.imread(path.get())
    p=(255 - img)
    cv2.imshow("negative",p)
    cv2.waitKey(0)
    return p
    

    
    
    
def gray_img():
    img=cv2.imread(path.get(),0)
    cv2.imshow("gray",img)
    cv2.waitKey(0)
    return img






def bluer_img():
    img=cv2.imread(path.get())
    bluer=cv2.blur(img,(9,9))
    cv2.imshow("bluer Image",bluer)
    cv2.waitKey(0)
    return bluer




def Gaussian_img():
    img=cv2.imread(path.get())
    bluer=cv2.GaussianBlur(img,(9,9),0)
    cv2.imshow("Gaussian image",bluer)
    cv2.waitKey(0)
    return bluer




def edge_img():
    img=cv2.imread(path.get())
    cane=cv2.Canny(img,10,50)
    cv2.imshow("edge canny",cane)
    cv2.waitKey(0)
    return img



def laplacian_img():
    img=cv2.imread(path.get())
    laplacian = cv2.Laplacian(img,cv2.CV_64F)
    cv2.imshow("laplacianFilter",laplacian)
    cv2.waitKey(0)
    return laplacian


def soboul_img():
    img=cv2.imread(path.get())
    kernal=np.array(
        [
            [0,1,0],
            [1,4,1],
            [0,1,0]
        ]
        )
    sobel = cv2.Sobel(img, cv2.CV_64F, 0, 1,kernal)
    cv2.imshow("sobel",sobel)
    cv2.waitKey(0)
    return sobel



def pyramid_img():
    src = cv2.imread(path.get())
    img1=cv2.pyrDown(src)
    img2=cv2.pyrDown(img1)
    cv2.imshow("pyramid",img1)
    cv2.imshow("pyramid",img2)
    cv2.waitKey(0)
    return img1 ,img2


def minimize_img():
    img=cv2.imread(path.get())
    height,width,_=img.shape
    p=cv2.resize(img,(height*2,width*2))
    cv2.imshow("miniumize",p)
    cv2.waitKey(0)
    return p


def smooth_img():
    img=cv2.imread(path.get())
    kernal=np.ones((7,7),np.float32)/49
    des=cv2.filter2D(img,-1,kernal)
    cv2.imshow("smoothing",des)
    cv2.waitKey(0)
    return des


def sharp_img():
    img=cv2.imread(path.get())
    kernal=np.array([
    [-1,-1,-1],
    [-1,9,-1],
    [-1,-1,-1]
    ])
    sharp=cv2.filter2D(img,-1,kernal)
    cv2.imshow("sharping",sharp)
    cv2.waitKey(0)
    return sharp





def rgb_img():
    img=cv2.imread(path.get())
    p=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    cv2.imshow("RGB",p)
    cv2.waitKey(0)


    
    
def rotate_img():
    img=cv2.imread(path.get())
    height,width,_=img.shape
    p=cv2.getRotationMatrix2D((height/2,width/2),90,1)
    new=cv2.warpAffine(img,p,(height,width))
    cv2.imshow("rotate",new)
    cv2.waitKey(0)
    return new



def binary_img():
    img=cv2.imread(path.get())
    ret,thr=cv2.threshold(img,200,255,cv2.THRESH_BINARY)
    cv2.imshow("Binary",thr)
    cv2.waitKey(0)
    return thr



def tozero_img():
    img=cv2.imread(path.get())
    ret,thr=cv2.threshold(img,200,255,cv2.THRESH_TOZERO)
    cv2.imshow("tozero",thr)
    cv2.waitKey(0)
    return thr



def cartoon_img():
    img=cv2.imread(path.get())
    grey=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    grey=cv2.medianBlur(grey,5)
    edges=cv2.adaptiveThreshold(grey,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,9,9)
    cartoon=cv2.bilateralFilter(img,9,25,25)
    cartoon2=cv2.bitwise_and(cartoon,cartoon,mask=edges)
    cv2.imshow("cartoon",cartoon2)
    cv2.waitKey(0)
    return cartoon2




def crop_img():
    img=cv2.imread(path.get())
    height,width,_=img.shape
    crop=img[0:(int (height/2)),0:(int (width/2))]
    cv2.imshow("crop",crop)
    cv2.waitKey(0)
    return crop







# def face_img():
#     image = cv2.imread(path.get())
#     result_image = image.copy()
#     face_cascade_name = "./haarcascade_frontalface_alt.xml"
#     face_cascade = cv2.CascadeClassifier()
#     face_cascade.load(face_cascade_name)
#     grayimg = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
#     grayimg = cv2.equalizeHist(grayimg)
#     faces = face_cascade.detectMultiScale(grayimg, 1.1, 2, 0|cv2.cv.CV_HAAR_SCALE_IMAGE, (30, 30))
#     if len(faces) != 0:
#         for f in faces:         
#             x, y, w, h = [ v for v in f ]
#     cv2.rectangle(image, (x,y), (x+w,y+h), (255,255,0), 5)
#     sub_face = image[y:y+h, x:x+w]
#     sub_face = cv2.GaussianBlur(sub_face,(23, 23), 30)
#     result_image[y:y+sub_face.shape[0], x:x+sub_face.shape[1]] = sub_face
       

#     cv2.imshow("Detected face", result_image)
    















    
main=Frame(root)
main.pack(fill=BOTH,expand=1)



can=Canvas(main)
can.pack(side=LEFT,fill=BOTH,expand=1)





bg=PhotoImage(file="C:\\Users\\HP\\Pictures\\filters image.png")

label2=Label(can,image=bg)
label2.place_configure(x=0,y=0,relwidth=1,relheight=1)



scrol=Scrollbar(main,orient=VERTICAL,command=can.yview)
scrol.pack(side=RIGHT,fill=Y)



can.configure(yscrollcommand=scrol.set)
can.bind('<Configure>',lambda e : can.configure(scrollregion=can.bbox("all")))

second_frame=Frame(can,bg='#CD8162')

can.create_window((0,0),window=second_frame,anchor="nw")

path=Entry(can,width=30)
path.pack()
path.insert(0, "coppied the path of ur image")



ok=Button(can,text="show",command=read_img)
ok.pack()



def reset(*args):
    path.delete(0, END)

path.bind("<Button-1>", reset)





negativeg=Button(second_frame,text="negative",command=neg_img).grid(row=0,column=0,pady=10,padx=10)

gray=Button(second_frame,text="gray",command=gray_img).grid(row=1,column=0,pady=10,padx=10)

bluer=Button(second_frame,text="bluering",command=bluer_img).grid(row=2,column=0,pady=10,padx=10)

gaussian=Button(second_frame,text="Gaussian",command=Gaussian_img).grid(row=3,column=0,pady=10,padx=10)

edge=Button(second_frame,text="Edge",command=edge_img).grid(row=4,column=0,pady=10,padx=10)

Laplacian=Button(second_frame,text="laplacian",command=laplacian_img).grid(row=5,column=0,pady=10,padx=10)

rotate=Button(second_frame,text="Rotate",command=rotate_img).grid(row=6,column=0,pady=10,padx=10)

RGB=Button(second_frame,text="RGB",command=rgb_img).grid(row=7,column=0,pady=10,padx=10)

maxmize=Button(second_frame,text="maxmize",command=minimize_img).grid(row=8,column=0,pady=10,padx=10)

Binary=Button(second_frame,text="Binary",command=binary_img).grid(row=9,column=0,pady=10,padx=10)

sharp=Button(second_frame,text="sharping",command=sharp_img).grid(row=10,column=0,pady=10,padx=10)

smooth=Button(second_frame,text="smothing",command=smooth_img).grid(row=11,column=0,pady=10,padx=10)

sobol=Button(second_frame,text="sobol",command=soboul_img).grid(row=12,column=0,pady=10,padx=10)

ToZero=Button(second_frame,text="Tozero",command=tozero_img).grid(row=13,column=0,pady=10,padx=10)

cartoon=Button(second_frame,text="Cartoon",command=cartoon_img).grid(row=14,column=0,pady=10,padx=10)

crop=Button(second_frame,text="crop",command=crop_img).grid(row=15,column=0,pady=10,padx=10)

pyramid=Button(second_frame,text="pyramid",command=pyramid_img).grid(row=16,column=0,pady=10,padx=10)

# face=Button(second_frame,text="Face",command=face_img).grid(row=17,column=0,pady=10,padx=10)

















root.mainloop()
```
